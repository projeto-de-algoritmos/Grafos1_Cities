# Aplicando busca em largura - Breadth First Search (BFS)

```python
def backtrace(parent, start, end):
    path = [end]
    while path[-1] != start:
        path.append(parent[path[-1]])
    path.reverse()
    return path

def bfs(graph, start, end):
    visited = defaultdict()
    parent = {}
    queue = []
    visited.clear()
    visited[start] = True
    queue.append(start)
    while queue:
        node = queue.pop(0)
        if node == end:
            return backtrace(parent, start, end)
        for adjacent in graph.get(node, []):
                if adjacent not in visited.keys():
                    if node not in queue :
                        parent[adjacent] = node # <<<<< record its parent 
                        queue.append(adjacent)
                        visited[adjacent] = True
# country_path = bfs(graph, 'Brazil', 'Canada')
# if country_path:
#     resultant_tree = nx.Graph()
#     i, j = 0, 1
#     while j < len(country_path):
#         resultant_tree.add_edge(country_path[i], country_path[j])
#         i += 1
#         j += 1
#     file = open("data/happyTourist.png", "rb")
#     image = file.read()
#     resultant_image = widgets.Image(
#                                 value=image,
#                                 format='png',
#                                 width=300,
#                                 height=400,
#                             )
# else:
#     file = open("data/sadTourist.png", "rb")
#     image = file.read()
#     resultant_image = widgets.Image(
#                                 value=image,
#                                 format='png',
#                                 width=300,
#                                 height=400,
#                             )
```

## Utilizando jupyter widgets

```python
def on_click_bfs(start, end):
    country_path = bfs(graph, start, end)
    print("Início: ", start,"\nFim:", end)
    resultant_tree = nx.Graph()
    print("\n",country_path)
    if country_path and len(country_path)>1:
        i, j = 0, 1
        while j < len(country_path):
            resultant_tree.add_edge(country_path[i], country_path[j])
            i += 1
            j += 1
        file = open("data/happyTourist.png", "rb")
        image = file.read()
        resultant_image = widgets.Image(
                                    value=image,
                                    format='png',
                                    width=300,
                                    height=400,
                                )
    else:
        file = open("data/sadTourist.png", "rb")
        image = file.read()
        resultant_image = widgets.Image(
                                    value=image,
                                    format='png',
                                    width=300,
                                    height=400,
                                )
    if country_path and len(country_path)>1:
        display(resultant_image)
        nx.draw_networkx(resultant_tree)
    else:
        display(resultant_image)
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: interact_manual(on_click_bfs, start= start_dropdown_widget,end= end_dropdown_widget)
```

```python
def continent_graph(continent):
    continent_df = run.data.df_border[run.data.df_border['continent'] == continent]
    graph = defaultdict(list)
    for item in continent_df.country_name.unique():
        graph[item] = run.data.df_border.border_list[run.data.df_border.country_name==item].item()
    continent_graph = nx.Graph()
    continent_graph.add_nodes_from(list(graph.keys()))
    for country in list(graph.keys()):
        for frontier in graph[country]:
            continent_graph.add_edge(country,frontier)
    return continent_graph
```

## Visualização de grafo dos continentes

```python
asia = continent_graph('ASIA')
europe = continent_graph('EUROPE')
america = continent_graph('AFRICA')
oceania = continent_graph('OCEANIA')
africa = continent_graph('AFRICA')
```

```python
f, axs = plt.subplots(5,1,figsize=(15,15))

plt.subplot(221)
nx.draw_networkx(america)
plt.title('America', fontsize=30)

plt.subplot(222)
nx.draw_networkx(asia)
plt.title('Asia', fontsize=30)

plt.subplot(223)
nx.draw_networkx(africa)
plt.title('Africa', fontsize=30)

plt.subplot(224)
plt.title('Oceania', fontsize=30)
nx.draw_networkx(oceania)
```
